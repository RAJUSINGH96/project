```python
import sqlite3
from tkinter import *
from tkinter.ttk import Combobox
from datetime import datetime
import time
from tkinter import messagebox
from tkinter.ttk import Style,Treeview,Scrollbar

try:
    conobj=sqlite3.connect(database="banking.sqlite")
    curobj=conobj.cursor()
    curobj.execute("create table accounts(acn integer primary key autoincrement,name text,pass text,email text,mob text,bal float,type text,opendate text)")
    curobj.execute("create table txns(acn int,amt float,updatebal float,type text,txndate text)") 
    conobj.close()
    print("tables created")
except:
    print("somthing went wrong,might be tables already exists") 

win=Tk()
win.state("zoomed")
win.configure(bg="powder blue")
win.resizable(width=False,height=False)

title=Label(win,text="Banking Automation",font=('arial',60,"bold"),bg="powder blue")
title.pack() 

date=Label(win,text=f"{datetime.now().date()}",font=('arial',15,"bold"),bg="powder blue")
date.place(relx=.9,rely=.1) 

def mainscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="pink")

    def new():
        frm.destroy()
        newscreen()

    def fp():
        frm.destroy()
        fpscreen()

    def reset():
        e_acn.delete(0,"end")
        e_pass.delete(0,"end")
        e_acn.focus()

    def login():
        acn=e_acn.get()
        pwd=e_pass.get()

        if(len(acn)==0 or len(pwd)==0):
            messagebox.showerror("Login","Empty fields are not allowed!")
        else:
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=? and pass=?",(acn,pwd))
            tup=curobj.fetchone()
            if tup==None:
                messagebox.showerror("Login","Invalid ACN/PASS")
            else:
                global uname,uacn 
                uacn=tup[0] 
                uname=tup[1] 
                frm.destroy()
                homescreen() 

    lbl_acn=Label(frm,text="ACN",font=("arial",20,"bold"),bg="pink")
    lbl_acn.place(relx=.3,rely=.1)

    e_acn=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_acn.place(relx=.4,rely=.1)
    e_acn.focus()

    lbl_pass=Label(frm,text="pass",font=("arial",20,"bold"),bg="pink")
    lbl_pass.place(relx=.3,rely=.2) 

    e_pass=Entry(frm,font=("arial",20,"bold"),bd=2,show="*")
    e_pass.place(relx=.4,rely=.2)

    lgn_btn=Button(frm,text="Login",command=login,font=("arial",20,"bold"),bd=2)
    lgn_btn.place(relx=.42,rely=.3)

    reset_btn=Button(frm,text="Reset",font=("arial",20,"bold"),bd=2,command=reset)
    reset_btn.place(relx=.5,rely=.3)

    fp_btn=Button(frm,text="forgot password",command=fp,font=("arial",20,"bold"),bd=2,width=16)
    fp_btn.place(relx=.4,rely=.4)

    new_btn=Button(frm,text="Create new account",font=("arial",20,"bold"),bd=2,width=20,command=new)
    new_btn.place(relx=.38,rely=.5)  
 
def newscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="pink")

    def back():
        frm.destroy()
        mainscreen()

    def reset():
        e_name.delete(0,"end")
        e_pass.delete(0,"end")
        e_mob.delete(0,"end") 
        e_name.focus() 

    def openacndb():
        bal=0
        opendate=time.ctime()
        name=e_name.get()
        pwd=e_pass.get()
        email=e_eamil.get()
        mob=e_mob.get()
        acntype=cb_type.get()

        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor() 
        curobj.execute("insert into accounts(name,pass,email,mob,bal,type,opendate) values(?,?,?,?,?,?,?)",(name,pwd,email,mob,bal,acntype,opendate))
        conobj.commit()
        curobj.close()

        curobj=conobj.cursor() 
        curobj.execute("select max(acn) from accounts")
        tup=curobj.fetchone()
        messagebox.showinfo("New Account",f"Account opened with ACN:{tup[0]}")
        conobj.close() 



    lbl_name=Label(frm,text="Name",font=("arial",20,"bold"),bg="pink")
    lbl_name.place(relx=.3,rely=.1)

    e_name=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_name.place(relx=.4,rely=.1)
    e_name.focus()

    lbl_pass=Label(frm,text="Password",font=("arial",20,"bold"),bg="pink")
    lbl_pass.place(relx=.3,rely=.2)

    e_pass=Entry(frm,font=("arial",20,"bold"),bd=2,show="*")
    e_pass.place(relx=.4,rely=.2)

    lbl_email=Label(frm,text="Email",font=("arial",20,"bold"),bg="pink")
    lbl_email.place(relx=.3,rely=.3)

    e_eamil=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_eamil.place(relx=.4,rely=.3)

    lbl_mob=Label(frm,text="Mobil no",font=("arial",20,"bold"),bg="pink")
    lbl_mob.place(relx=.3,rely=.4)

    e_mob=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_mob.place(relx=.4,rely=.4)

    lbl_type=Label(frm,text="Type",font=("arial",20,"bold"),bg="pink")
    lbl_type.place(relx=.3,rely=.5) 

    cb_type=Combobox(frm,values=["Saving","Current"],font=("arial",20,"bold"))
    cb_type.current(0)
    cb_type.place(relx=.4,rely=.5)

    open_btn=Button(frm,text="open",font=("arial",20,"bold"),bd=2,command=openacndb)
    open_btn.place(relx=.42,rely=.6)

    reset_btn=Button(frm,text="Reset",font=("arial",20,"bold"),bd=2,command=reset) 
    reset_btn.place(relx=.5,rely=.6)

    back_btn=Button(frm,text="Back",font=("arial",20,"bold"),bd=2,command=back)  
    back_btn.place(relx=0,rely=0)

def fpscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="pink")

    def back():
        frm.destroy()
        mainscreen()

    def recpassdb():
        acn=e_acn.get()
        email=e_email.get()
        mob=e_mob.get()

        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select pass from accounts where acn=? and email=? and mob=?",(acn,email,mob))
        tup=curobj.fetchone() 
        if tup==None:
            messagebox.showerror("Forgot Pass","Account does not exist") 
        else:
            messagebox.showinfo("Forgot Pass",f"Your Pass:{tup[0]}")

    back_btn=Button(frm,text="back",font=("arial",20,"bold"),bd=2,command=back)
    back_btn.place(relx=0,rely=0)

    lbl_acn=Label(frm,text="Acn",font=("arial",20,"bold"),bg="pink")
    lbl_acn.place(relx=.3,rely=.2)

    e_acn=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_acn.place(relx=.4,rely=.2)
    e_acn.focus()
    
    lbl_email=Label(frm,text="Email",font=("arial",20,"bold"),bg="pink")
    lbl_email.place(relx=.3,rely=.3)

    e_email=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_email.place(relx=.4,rely=.3)

    lbl_mob=Label(frm,text="Mob",font=("arial",20,"bold"),bg="pink")
    lbl_mob.place(relx=.3,rely=.4)

    e_mob=Entry(frm,font=("arial",20,"bold"),bd=2)
    e_mob.place(relx=.4,rely=.4)

    rec_btn=Button(frm,text="recover",font=("arial",20,"bold"),bd=2,command=recpassdb)
    rec_btn.place(relx=.42,rely=.5)

    reset_btn=Button(frm,text="reset",font=("arial",20,"bold"))
    reset_btn.place(relx=.53,rely=.5)

def homescreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg="pink")

    def logout():
        frm.destroy()
        mainscreen()

    def detailsscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white") 

        lbl=Label(ifrm,text="This id Account Details Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack()

        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,))
        tup=curobj.fetchone()

        lbl_acn=Label(ifrm,text=f"Account No : {tup[0]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.2)

        lbl_acn=Label(ifrm,text=f"Name : {tup[1]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.3)

        lbl_acn=Label(ifrm,text=f"Account Email : {tup[3]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.4)

        lbl_acn=Label(ifrm,text=f"mob : {tup[4]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.5)

        lbl_acn=Label(ifrm,text=f"Account Bal : {tup[5]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.6)

        lbl_acn=Label(ifrm,text=f"Account Type : {tup[6]}",font=("arial",15),bg="white")
        lbl_acn.place(relx=.2,rely=.7)

        lbl_acn=Label(ifrm,text=f"Account opendate : {tup[7]}",font=("arial",15),bg="white") 
        lbl_acn.place(relx=.2,rely=.8)


    def depositscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white") 

        def depdb():
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            curobj=conobj.cursor() 
            curobj.execute("update accounts set bal=bal+? where acn=?",(amt,uacn))
            curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal+amt,"cr.",time.ctime()))
            conobj.commit()
            conobj.close()

            messagebox.showinfo("Deposite",f"Amount deposited,updated bal:{bal+amt}") 

        lbl=Label(ifrm,text="This id Deposit Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack()

        lbl_amt=Label(ifrm,text="Amount",font=("arial",15,"bold"),bg="white")
        lbl_amt.place(relx=.2,rely=.3)

        e_amt=Entry(ifrm,font=("arial",20,"bold"),bg="white",bd=2)
        e_amt.place(relx=.35,rely=.3)

        dep_btn=Button(ifrm,command=depdb,text="Deposit",font=("arial",15,"bold"),bd=2)
        dep_btn.place(relx=.4,rely=.4)

    def withdrawscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white")

        def withdrawdb(): 
            amt=float(e_amt.get())
            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select bal from accounts where acn=?",(uacn,))
            bal=curobj.fetchone()[0]
            curobj.close()
            if bal>amt:
                curobj=conobj.cursor() 
                curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal-amt,"db.",time.ctime()))
                conobj.commit()

                messagebox.showinfo("Withdraw",f"Amount withdrawn,updated bal:{bal-amt}")
            else:
                messagebox.showwarning("withdraw",f"Insufficient Bal:{bal}")  
            
            conobj.close() 

        lbl=Label(ifrm,text="This id withdraw Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack()

        lbl_amt=Label(ifrm,text="Amount",font=("arial",15,"bold"),bg="white")
        lbl_amt.place(relx=.2,rely=.3)

        e_amt=Entry(ifrm,font=("arial",20,"bold"),bg="white",bd=2)
        e_amt.place(relx=.35,rely=.3)

        dep_btn=Button(ifrm,command=withdrawdb,text="Withdraw",font=("arial",15,"bold"),bd=2)
        dep_btn.place(relx=.4,rely=.4)


    def profilscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white")

        def update():
            name=e_name.get()
            pwd=e_pass.get() 
            email=e_email.get()
            mob=e_mob.get() 
            conobj=sqlite3.connect(database="banking.sqlite") 
            curobj=conobj.cursor()
            curobj.execute("update accounts set name=?,pass=?,email=?,mob=? where acn=?",(name,pwd,email,mob,uacn)) 
            conobj.commit()
            conobj.close()
            messagebox.showinfo("update","Profile Updated") 
            ifrm.destroy()
            global uname
            profilscreen

        lbl=Label(ifrm,text="This id profil detail Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack() 

        lbl_name=Label(ifrm,text="Name",font=("arial",15,"bold"),bg="white")
        lbl_name.place(relx=.1,rely=.2) 

        e_name=Entry(ifrm,font=("arial",15,"bold"),bg="white",bd=2)
        e_name.place(relx=.1,rely=.28)

        lbl_pass=Label(ifrm,text="pass",font=("arial",15,"bold"),bg="white") 
        lbl_pass.place(relx=.1,rely=.4)

        e_pass=Entry(ifrm,font=("arial",15,"bold"),bg="white",bd=2)
        e_pass.place(relx=.1,rely=.48)

        lbl_email=Label(ifrm,text="Email",font=("arial",15,"bold"),bg="white")
        lbl_email.place(relx=.5,rely=.2) 

        e_email=Entry(ifrm,font=("arial",15,"bold"),bg="white",bd=2)
        e_email.place(relx=.5,rely=.28)

        lbl_mob=Label(ifrm,text="Mob",font=("arial",15,"bold"),bg="white") 
        lbl_mob.place(relx=.5,rely=.4) 

        e_mob=Entry(ifrm,font=("arial",15,"bold"),bg="white",bd=2)
        e_mob.place(relx=.5,rely=.48)  

        dep_btn=Button(ifrm,text="Update",font=("arial",15,"bold"),bd=2,command=update)
        dep_btn.place(relx=.5,rely=.6) 

        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from accounts where acn=?",(uacn,)) 
        tup=curobj.fetchone() 
        curobj.close() 

        e_name.insert(0,tup[1]) 
        e_pass.insert(0,tup[2]) 
        e_email.insert(0,tup[3]) 
        e_mob.insert(0,tup[4]) 

    def historyscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white") 

        lbl=Label(ifrm,text="This id History Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack() 

        tv=Treeview(ifrm)
        tv.place(x=60,y=100,height=100,width=600)

        style=Style() 
        style.configure("Treeview.Heading",font=('Arial',10,'bold'),foreground='brown') 

        sb=Scrollbar(ifrm,orient="vertical",command=tv.yview)
        sb.place(x=650,y=100,height=100)
        tv.configure(yscrollcommand=sb.set) 

        tv['columns']=('date','amt','type','upbal') 
        tv.column('date',width=80,anchor='c')
        tv.column('amt',width=40,anchor='c')
        tv.column('type',width=40,anchor='c')
        tv.column('upbal',width=40,anchor='c') 

        tv.heading('date',text='Txn Date')
        tv.heading('amt',text='Amount')
        tv.heading('type',text='Txn type')
        tv.heading('upbal',text='Updated Bal') 

        tv['show']='headings'

        conobj=sqlite3.connect(database="banking.sqlite")
        curobj=conobj.cursor()
        curobj.execute("select * from txns where acn=?",(uacn,))

        for row in curobj:
            tv.insert("","end",values=(row[4],row[1],row[3],row[2])) 

    def transferscreen():
        ifrm=Frame(frm,highlightbackground="black",highlightthickness=2)
        ifrm.place(relx=.3,rely=.15,relwidth=.5,relheight=.6)
        ifrm.configure(bg="white") 

        def Transfer():
            to=e_to.get()
            amt=float(e_amt.get()) 

            conobj=sqlite3.connect(database="banking.sqlite")
            curobj=conobj.cursor()
            curobj.execute("select * from accounts where acn=?",(to,)) 
            tup=curobj.fetchone() 
            curobj.close() 
            if tup==None: 
                messagebox.showerror("Transfer","To ACN does not exist")
            else:
                curobj=conobj.cursor() 
                curobj.execute("select bal from accounts where acn=?",(uacn,))
                bal=curobj.fetchone()[0] 
                if bal>=amt: 
                    curobj=conobj.cursor()
                    curobj.execute("update accounts set bal=bal+? where acn=?",(amt,to))
                    curobj.execute("update accounts set bal=bal-? where acn=?",(amt,uacn))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(to,amt,bal+amt,"cr.",time.ctime()))
                    curobj.execute("insert into txns values(?,?,?,?,?)",(uacn,amt,bal-amt,"Db.",time.ctime()))
                    conobj.commit()
                    conobj.close()
                    messagebox.showinfo("Transfer",f"Rs. {amt} is transfered to {to} ACN") 

                else:
                    messagebox.showinfo("Transfer",f"Insufficient Bal:{bal}") 

        lbl=Label(ifrm,text="This id Transfer Screen",font=("arial",20,"bold"),bg="white")
        lbl.pack()

        lbl_to=Label(ifrm,text="To ACN",font=("arial",15,"bold"),bg="white")
        lbl_to.place(relx=.2,rely=.3)  

        e_to=Entry(ifrm,font=("arial",20,"bold"),bg="white",bd=2)
        e_to.place(relx=.35,rely=.3) 

        lbl_amt=Label(ifrm,text="Amount",font=("arial",15,"bold"),bg="white")
        lbl_amt.place(relx=.2,rely=.4) 

        e_amt=Entry(ifrm,font=("arial",20,"bold"),bg="white",bd=2)
        e_amt.place(relx=.35,rely=.4)

        btn=Button(ifrm,command=Transfer,text="Transfer",font=("arial",15,"bold"),bd=2)
        btn.place(relx=.4,rely=.5)         

    lbl_wel=Label(frm,text=f"welcome,{uname}",font=("arial",20,"bold"),bg="pink")
    lbl_wel.place(relx=0,rely=0)

    logout_btn=Button(frm,text="logout",command=logout,font=("arial",20,"bold"),bg="pink")
    logout_btn.place(relx=.92,rely=0)

    details_btn=Button(frm,text="details",command=detailsscreen,font=("arial",25,"bold"),bg="pink",width=12)
    details_btn.place(relx=0,rely=.1)

    deposit_btn=Button(frm,text="deposit",command=depositscreen,font=("arial",25,"bold"),bg="pink",width=12)
    deposit_btn.place(relx=0,rely=.2)

    withdraw_btn=Button(frm,text="withdraw",command=withdrawscreen,font=("arial",25,"bold"),bg="pink",width=12)
    withdraw_btn.place(relx=0,rely=.3)

    profile_btn=Button(frm,text="Update profile",command=profilscreen,font=("arial",25,"bold"),bg="pink",width=12)
    profile_btn.place(relx=0,rely=.4)

    history_btn=Button(frm,text="History",command=historyscreen,font=("arial",25,"bold"),bg="pink",width=12)
    history_btn.place(relx=0,rely=.5) 

    trans_btn=Button(frm,text="Transfer",command=transferscreen,font=("arial",25,"bold"),bg="pink",width=12)
    trans_btn.place(relx=0,rely=.6) 

mainscreen()
win.mainloop()
```
