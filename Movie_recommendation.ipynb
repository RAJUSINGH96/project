```python
import pandas as pd
from tkinter import * 
from datetime import datetime
from tkinter.ttk import Combobox
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors
from tkinter import messagebox

df=pd.read_csv("c:/users/raju/downloads/movies_content.csv")
df=df[['name','description']]
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
df.reset_index(drop=True,inplace=True)
df.name=df.name.map(str.lower)
df1=df.name.values
tv=TfidfVectorizer() 
matrix=tv.fit_transform(df.description).toarray()
model=NearestNeighbors(metric='cosine')
model.fit(matrix)

win=Tk()
win.state("zoomed")
win.configure(bg="pink")

title=Label(win,text="Movie Recommendation",font=("arial",30,"bold"),bg="pink",fg="black")
title.place(relx=.35,rely=.04)

date=Label(win,text=f"{datetime.now().date()}",font=('arial',20,'bold'),bg="pink",fg="black")
date.place(relx=.8,rely=.1)

def home_screen():
    frm=Frame(win,bg='sky blue') 
    frm.place(relx=0,rely=.15,relwidth=1,relheight=1)

    def login(entry_user=None,entry_pass=None):
        if(entry_user!=None and entry_pass!=None):
            user=entry_user.get()
            pwd=entry_pass.get() 
        else:
            user="admin"
            pwd="admin"
        if(len(user)==0 or len(pwd)==0):
            messagebox.showwarning("validation","Please fill both fields")
            return
        else:
            if(user=="admin" or pwd=="admin"):
                frm.destroy()
                mainscreen()            

    lbl_user=Label(frm,text="Username",font=('',20,'bold'),bg='sky blue')
    lbl_user.place(relx=.28,rely=.3)

    entry_user=Entry(frm,font=('',20,'bold'),bd=2)
    entry_user.place(relx=.42,rely=.3)
    entry_user.focus()

    lbl_pass=Label(frm,text="Password",font=('',20,'bold'),bg='sky blue')
    lbl_pass.place(relx=.28,rely=.4)

    entry_pass=Entry(frm,font=('',20,'bold'),bd=2,show="*")
    entry_pass.place(relx=.42,rely=.4)
        
    btn_login=Button(frm,command=login,text="login",font=('',20,'bold'),bd=2,width=10)
    btn_login.place(relx=.45,rely=.5)    

def mainscreen():
    frm=Frame(win)
    frm.place(relx=0,rely=.15,relwidth=1,relheight=.85)
    frm.configure(bg='sky blue')

    def prediction(): 
        ifrm=Frame(frm,highlightbackground='black',highlightthickness=2)
        ifrm.place(relx=.2,rely=.4,relwidth=.26,relheight=.35)
        ifrm.configure(bg='white') 

        lbl=Label(ifrm,text="Similar movie recommended for you",font=('arial',15,'bold'),bg='white',fg="green")
        lbl.pack() 

        movie_name=cb_type.get()
        mv=movie_name.lower()
        mi=df[df.name==mv].index[0]
        dis,idx=model.kneighbors([matrix[mi]],n_neighbors=5)
        v=0
        for i in idx[0][1:]:
            l=Label(ifrm,text=df.loc[i].values[0],font=('',15,'bold'),bg='white',fg='maroon')
            l.place(x=50,y=50+v)
            v+=30

    cb_type=Combobox(win,values=list(df1),font=("arial",30,'bold')) 
    cb_type.current(0)
    cb_type.place(relx=.26,rely=.3)

    btn_login=Button(win,command=prediction,text="Recommendation",font=('',20,'bold'),bd=10,width=15)
    btn_login.place(relx=.47,rely=.46)

home_screen()
win.mainloop()
```
